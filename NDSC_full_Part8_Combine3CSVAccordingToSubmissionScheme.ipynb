```python
# merge the 3 files produces in order of beauty, fashion, mobile
# beautyTest  is if ran Part7
# beautyTestKneigh is if ran Part10
import pandas as pd
# from part 7 output
dfbeauty = pd.read_csv('beautyTest.csv') # import training set
dffashion = pd.read_csv('fashionTest.csv') # import training set
dfmobile = pd.read_csv('mobileTest.csv') # import training set
df3Grps = pd.concat([dfbeauty,dffashion, dfmobile])
print(len(df3Grps))


```

```python
# remove unneeded columns
df3Grps1 = df3Grps[['itemid','predicted']]



# rename the column to the right column header
df3Grps2 =df3Grps1.rename(columns={"predicted":"Category"}) 
df3Grps2.set_index(['itemid'])
print(len(df3Grps2))



```

```python
# read to a csv as output to be uploaded to Kaggle
# https://stackoverflow.com/questions/20845213/how-to-avoid-python-pandas-creating-an-index-in-a-saved-csv
df3Grps2.to_csv('[Beginner]_Final _Submission-(RevengeforTheOstracised).csv', index=False)

```
